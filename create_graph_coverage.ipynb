```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import patches
import math 
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
SAVE_DIRECTORY = 'csv/'
```

```python
# coverage_ratio
df = pd.read_csv(SAVE_DIRECTORY + 'coverage_ratio.csv')
```

```python
x = df['time'][1:]
y = df['coverage_ratio'][1:]

plt.plot(x, y, '-', linewidth=0.5)
plt.xlim(0,700)  # csvファイルに合わせて手入力で調整
plt.ylim(0, 100)
plt.xlabel('time [s]')
plt.ylabel('coverage ratio [%]')
plt.title('coverage ratio by number of times')
plt.show()
```
